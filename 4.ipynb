```python
inp_file = open('4.txt', 'r')
inp = inp_file.read()
lines = inp.split('\n')[:-1]
```

# Part 1

```python
letters = lambda l: ''.join(l.split('-')[:-1])
tail = lambda l: l.split('-')[-1]
get_id = lambda l: int(tail(l).split('[')[0])
get_check = lambda l: tail(l).split('[')[1].strip(']')
```

```python
def get_sorted_counts(line):
    let = letters(line)
    char_list = sorted(list(set(let)))
    counts = map(lambda c: let.count(c), char_list)
    sorted_charcounts = sorted(zip(char_list, counts), key=lambda x: x[1], reverse=True)
    return sorted_charcounts
    
def compute_checksum(line):
    sorted_letters = map(lambda x: x[0], get_sorted_counts(line)[:5])
    comp_checksum = ''.join(sorted_letters)
    return comp_checksum

def test_checksum(line):
    return compute_checksum(line) == get_check(line)
```

```python
correct_sum = map(test_checksum, lines)
correct_ids = [get_id(l) for l, check in zip(lines, correct_sum) if check == True]
sum(correct_ids)
```

# Part 2

```python
def rotate_char(c, cipher):
    return chr(((ord(c)-ord('a') + cipher) % (ord('z')-ord('a')+1)) + ord('a'))

def decrypt_all(line):
    strings = line.split('-')[:-1]
    room_id = get_id(line)
    decrypt_one = lambda s: ''.join(map(lambda c: rotate_char(c, room_id), s))
    decrypted_strings = map(decrypt_one, strings)
    return ' '.join(decrypted_strings) + ' ' + str(room_id)
```

```python
map(decrypt_all, lines)
```
