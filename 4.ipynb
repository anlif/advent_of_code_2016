```python
letters = lambda l: ''.join(l.split('-')[:-1])
tail = lambda l: l.split('-')[-1]
get_id = lambda l: int(tail(l).split('[')[0])
get_check = lambda l: tail(l).split('[')[1].strip(']')
```

```python
def get_sorted_counts(line):
    let = letters(line)
    char_list = sorted(list(set(let)))
    counts = map(lambda c: let.count(c), char_list)
    sorted_charcounts = sorted(zip(char_list, counts), key=lambda x: x[1], reverse=True)
    return sorted_charcounts
    
def compute_checksum(line):
    sorted_letters = map(lambda x: x[0], get_sorted_counts(line)[:5])
    comp_checksum = ''.join(sorted_letters)
    return comp_checksum

def test_checksum(line):
    return compute_checksum(line) == get_check(line)
```
